```python
import pandas as pd
from geopy.geocoders import Nominatim

# Function to get country name from city name using geocoding


def get_country_name(city):
    geolocator = Nominatim(user_agent="country_finder")
    location = geolocator.geocode(city)
    if location is not None:
        return location.address.split(",")[-1].strip()
    else:
        return None


# Read the CSV file into a Pandas DataFrame
csv_file = "AQI and Lat Long of Countries.csv"
df = pd.read_csv(csv_file)

# Create a list to store the updated country details
updated_details = []

# Iterate through the DataFrame to update missing country names
for index, row in df.iterrows():
    country = row["Country"]
    city = row["City"]
    if pd.isnull(country):  # Check if the country is missing
        new_country = get_country_name(city)
        if new_country:
            df.at[index, "Country"] = new_country
            updated_details.append(
                f"Updated country for city '{city}' to '{new_country}'")
            print(f"Updated country for city '{city}' to '{new_country}'")

# Save the updated DataFrame back to the CSV file
df.to_csv(csv_file, index=False)

# Save the updated details to a text file
text_file = "updated_details.txt"
with open(text_file, "w") as file:
    file.write("\n".join(updated_details))
```
