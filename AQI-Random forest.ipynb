```python
## classes are unbalanced. However without balancing them first lets do the classifications and get the accuracies. 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Assuming X and Y are pandas DataFrame objects
X = data[['CO AQI Value', 'Ozone AQI Value', 'NO2 AQI Value', 'PM2.5 AQI Value', 'SO2', 'Continent']]
Y = data['AQI Category']

# Convert the ordinal response variable to numerical values
Y_numerical = Y.map({'Good': 1, 'Moderate': 2, 'Unhealthy': 3})

# Convert numerical response into ordinal categorical type
Y_ordinal = pd.Categorical(Y_numerical, ordered=True)

# scale the predictor variables except for 'Continent'
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X.iloc[:, :-1])  # Exclude 'Continent'

# One-hot encode the 'Continent' variable
encoder = OneHotEncoder()
continent_encoded = encoder.fit_transform(X[['Continent']]).toarray()

# Get the names of the one-hot encoded columns
encoded_column_names = encoder.get_feature_names_out(['Continent'])

# Combine the one-hot encoded column names with the rest of the feature names
all_feature_names = X.columns.tolist()[:-1] + encoded_column_names.tolist()

# Combine scaled predictor variables and one-hot encoded 'Continent'
X_scaled = np.hstack((X_scaled, continent_encoded))

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y_ordinal, test_size=0.2, random_state=42)
```

```python
import pickle

# Save the scaler model
with open('scaler_model.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# Save the encoder model
with open('encoder_model.pkl', 'wb') as f:
    pickle.dump(encoder, f)
```

```python
## before fitting LDA we need to check the assumptions 
# (normality and homoscedasticity)

# check the normality assumption

from scipy.stats import shapiro
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Separate data into different ordinal response classes
X_classes = [X_train[Y_train == ordinal] for ordinal in np.unique(Y_train)]

# Perform Shapiro-Wilk test for normality on each class
alpha = 0.05
for ordinal, data in enumerate(X_classes, start=1):
    normality_results = []
    for feature_idx in range(X_train.shape[1]):
        _, p_value = shapiro(X_train[:, feature_idx])
        normality_results.append(p_value > alpha)
    
    print(f"Ordinal Response {ordinal} - Normality Test Results:", normality_results)

# True ---> normally distributed
# False ----> not normally distributed
```

```python
## Fit the LDA model
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda_model = LinearDiscriminantAnalysis()

# Fit the model on the scaled training data
lda_model.fit(X_train, Y_train)
```

```python
# Predict the classes on the test data
y_pred = lda_model.predict(X_test)

# Evaluate the model
accuracy = np.mean(y_pred == Y_test)
print("Accuracy:", accuracy)
```

```python
## Fit the QDA model

from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

# Instantiate the QDA model
qda_model = QuadraticDiscriminantAnalysis()

# Fit the model on the scaled training data
qda_model.fit(X_train, Y_train)
```

```python
# Predict the classes on the test data
y_pred = qda_model.predict(X_test)

# Evaluate the model
accuracy = np.mean(y_pred == Y_test)
print("Accuracy:", accuracy)
```

```python
## fit the ordinal logistic regression model

# Instantiate the ordinal logistic regression model
ordinal_model = LogisticRegression(multi_class='ovr', solver='liblinear', random_state=42)

# Fit the model on the scaled training data
ordinal_model.fit(X_train, Y_train)
```

```python
## Predict and evaluate

# Predict the ordinal classes on the test data
y_pred = ordinal_model.predict(X_test)

# Evaluate the model
accuracy = np.mean(y_pred == Y_test)
print("Accuracy:", accuracy)

mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Create a Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
rf_model.fit(X_train, Y_train)

# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, y_pred)
classification_rep = classification_report(Y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(classification_rep)
```
